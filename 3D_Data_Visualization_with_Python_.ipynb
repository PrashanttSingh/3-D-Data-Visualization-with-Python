```python
#import
import plotly.express as px
import pandas as pd
```

```python
## sample dataset
df3= pd.DataFrame({
    'month':['jan','feb','mar','apr','may','jun'],
     'sales':[100,200,300,400,500,600],
    'profit':[10,20,30,40,50,60]
})
```

```python
#line plot
fig=px.line(df3,x='month',y='sales',title='sales in diff. months')
fig.show()
```

```python
#barchart
fig=px.bar(df3,x='month',y='profit',title='profit in diff. months',color='month')

fig.show()
```

```python
#scatterplot
fig=px.scatter(df3,x='sales',y='profit',title='profit in diff. months',color='month')
fig.show()
```

```python
#histogram
fig=px.histogram(df3,x='sales',title='profit in diff. months',nbins=4)
fig.show()
```

```python
#boxplot
fig=px.box(df3,y='profit',title='profit in diff. months',)
fig.show()
```

```python
#import library
import pandas as pd
import plotly.express as px
```

```python
#read datset
df=pd.read_csv('https://github.com/YBIFoundation/Visualization/raw/main/RetailData.csv')
```

```python
daily=df.groupby('date',as_index=False)['revenue'].sum()
```

```python
daily.plot()
```

```python
daily['rev_7d']=daily['revenue'].rolling(7).mean()
```

```python
daily.plot()
```

```python
fig=px.line(daily,x='date',y=['revenue' ,'rev_7d'])
fig.show()
```

```python
monthly_cat=df.groupby(['month','category'],as_index=False)['revenue'].sum()
```

```python
fig=px.bar(monthly_cat,x='month',y='revenue',color='category',barmode='group')
fig.show()
```

```python
df
```

```python
revenue_region=df.groupby('region',as_index=False)['revenue'].sum()
```

```python
fig=px.pie(revenue_region,values='revenue',names='region',title='simple pie charts',color='revenue'

)
fig.show()
```

```python
product_sold=df.groupby(['units','category',] ,as_index=False)['unit_price'].mean()
```

```python
fig=px.scatter_3d(product_sold,x='category',y='unit_price',z='units',color='category')
fig.show()
```

```python
daily_region=df.groupby(['date','region'],as_index=False)['revenue'].sum()
```

```python
fig=px.line(daily_region,color='region',x='date',y='revenue',title='revenue across various months(slider enabled)'
,labels={'revenue':'Revenue()','date':'Date'})
fig.update_xaxes(rangeslider_visible=True,
                 rangeselector=dict(
                     buttons=list([
                         dict(count=1,label='1m',step='month',stepmode='backward'),
                         dict(count=3,label='3m',step='month',stepmode='backward'),
                         dict(count=5,label='5m',step='month',stepmode='backward'),
                         dict(step='all')
                     ]),
                     activecolor='red',
                     bgcolor="orange",
                      font=dict(color='darkblue', size=17)
                 ))

fig.show()
```
